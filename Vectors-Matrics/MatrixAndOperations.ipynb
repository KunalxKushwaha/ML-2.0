# Matrix Operations

## **1. Matrix Addition**

**Two Matrices can be added only if they have the same shape.**

```python
# Creating two matrices
A = np.array([[2,3],[4,5]])

B = np.array([[6,7],[8,9]])
print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
```

```python
A.shape
```

```python
B.shape
```

```python
# Adding the Matrix
sum = A + B
print(sum)
```

```python
# Taking a matrix with random values
C = np.random.randint(10,size=(3,3))
D = np.random.randint(10, size=(3,3))
print("Matrix C:")
print(C)
print("Matrix D:")
print(D)
```

```python
sum = C + D
#or sum = np.add(C,D)
print(sum)
```

## **2. Matrix Subtraction**

```python
# Creating two matrices
A = np.array([[2,3],[4,5]])

B = np.array([[6,7],[8,9]])
print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
```

```python
# Subtracting
diff = A - B
print(diff)
```

```python
# Taking a matrix with random values
C = np.random.randint(10,size=(3,3))
D = np.random.randint(20, size=(3,3))
print("Matrix C:")
print(C)
print("Matrix D:")
print(D)
```

```python
diff = np.subtract(C,D)
print(diff)
```

## **3. Multiplying a matrix with a Scalar**

```python
x = 5
y = np.random.randint(10,size=(4,4))
print(y)
```

```python
product = np.multiply(x,y)
print(product)
```

## **4. Multiplying 2 Matrices**

-->**The Number of column of first matrix should be equal to the number of rows to the second matrix**

```python
# Taking a matrix with random values
A = np.random.randint(5,size=(2,3))
B = np.random.randint(5, size=(3,4))
print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
```

```python
product = np.dot(A,B)
print(product)
```

```python
# Taking a matrix with random values
A = np.random.randint(5,size=(2,3))
B = np.random.randint(5, size=(2,4))
print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
```

## 5. Element Wise Multiplication

```python
# Taking a matrix with random values
A = np.random.randint(5,size=(3,3))
B = np.random.randint(5, size=(3,3))
print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
```

```python
product = np.multiply(A,B) # multiply element to element only like scalar.
print(product)
```
