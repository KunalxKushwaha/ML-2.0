# Importing Some Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
```

# Plotting a Vector

```python
plt.quiver(0,0,4,5)  # Quiver function is used to hold or Store the value of vectors. just like the holder quiver of arrows.
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles='xy', scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-3,-6, scale_units = 'xy', angles='xy', scale=1, color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## Addition of 2 Vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,5,1, scale_units = 'xy', angles='xy', scale=1, color='r')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## Subtraction of 2 Vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

diff = vector_1 - vector_2
print(diff)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,-1,5, scale_units = 'xy', angles='xy', scale=1, color='r')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## Multiply a Vector by Scalar

```python
vect_1 = np.asarray([0,0,2,3])
vect_2 = 2*vect_1
print(vect_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,4,6, scale_units = 'xy', angles='xy', scale=1, color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vect_1 = np.asarray([0,0,2,3])
vect_2 = -0.5*vect_1
print(vect_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-1,-1.5, scale_units = 'xy', angles='xy', scale=1, color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
