# Importing some Dependencies

```python
import numpy as np
```

## Creating a Matrix Using Numpy

```python
mat_1 = np.array([[2,3],[6,7]])
```

```python
print(mat_1)
```

```python
mat_1.shape
```

# Creating Matrices with Random Values

```python
random_mat = np.random.rand(3,3) # rand gives the random numbers between 0 and 1 only
print(random_mat)
```

```python
# Creating matrix with random Integers
rand_mat = np.random.randint(100, size=(4,5))  # randint gives the random values integer which are less than 100
print(rand_mat)
```

## Matrix with all the elements as 1

```python
mat_3 = np.ones((2,3))
print(mat_3)
```

```python
 mat_3 = np.ones((26,26), dtype=int) # dtype parameter defines the datatype of the values.
print(mat_3)
```

## Null or Zero Matrix

```python
null_mat = np.zeros((4,4), dtype=int)
print(null_mat)
```

## Identity Matrix

```python
ident_mat = np.eye(3,3)
print(ident_mat)
```

```python
ident_mat = np.eye(5,6)
print(ident_mat)
```

## Transpose of Matrix

```python
# Matrix with Random Integer Values
a = np.random.randint(100, size=(5,5))
print(a)
```

```python
transpose_a = np.transpose(a)
print(transpose_a)
```
