# Plotting a Vector

```python
plt.quiver(0,0,4,5)  # Quiver function is used to hold or Store the value of vectors. just like the holder quiver of arrows.
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles='xy', scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-3,-6, scale_units = 'xy', angles='xy', scale=1, color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
